```python
# 定义函数 

def devidePlus(m, n):               # devidePlus为随意命名，m、n表示变量
    return float(m)/n+ 1            # 给出这个函数的内涵，其中，float（m）表强制m转换为浮点数
```

```python
# for in 循环
# 案例一
for i in range(10): # 循环10次，从0开始，i是一个变量
    print i, i*10, i**2 # print打印，会自动换行
```

```python
# 案例二
for i in range(10):
    print devidePlus(i, 2) # 在前文定义了devideplus函数，m被替换为i，n被替换为2；（）表示一个元祖、数组
```

```python
# 案例三
r = [devidePlus(i, 2)  for i in range(10)] # 赋值，定义一个变量r，r为一个列表，r里面的（）为一个元祖、数组
r
```

```python
# map函数 map( func, seq1[, seq2...] )
map(devidePlus, [4, 2], [2, 1]) 

#（）为一个元祖、数组，放置参数 ； []为一个列表，里面表示m=4，n=2 
# 该函数是将func的命令作用于seq中的每一个元素，并用一个列表给出返回值 ；省时间的做法
```

```python
# while循环
# 案例一
j = 0
while j <10:
    print j  # 打印所有符合while的j
    j+=1 # avoid dead loop 防无限；该写法等价于j=j+1的写法
    
# 该循环的操作指：只要满足某种条件 就无限操作下去
```

```python
# 案例二
j = 0
while j <10:
    if j%2 != 0:     # ！= 指 不等于
        print j**2   # ** 指 j的二次方
    j+=1 # avoid dead loop
```

```python
# 案例三
j = 0
while j <50:
    if j == 30:
        break              # break 中断退出当前循环；contine 继续接下来的循环；return 返回，相当于退出循环，一般在方法中作用
    if j%2 != 0:           # j除以2的余数 不等于 0
        print j**2
    j+=1 # avoid dead loop
```
